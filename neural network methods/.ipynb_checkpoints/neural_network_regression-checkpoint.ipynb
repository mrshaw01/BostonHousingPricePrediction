## Import necessary modules

```python
import time
import random
import numpy as np
import pandas as pd
import seaborn as sns

import sklearn
import sklearn.model_selection

import torch
from torch import nn
```

## Load data

```python
df = pd.read_csv("../data/boston_house_prices_processed.csv")
df
```

## Scale data

```python
scaler = sklearn.preprocessing.StandardScaler()
df[df.columns] = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df
```

## Split data

```python
X = df.drop(columns="MEDV")
Y = df["MEDV"]
```

## Performance evaluation criteria

```python
def performance_evaluate(Y_true, Y_predicted):
    R2_score = sklearn.metrics.r2_score(Y_true, Y_predicted)
    MSE_score = sklearn.metrics.mean_squared_error(Y_true, Y_predicted)
    MAPE_score = sklearn.metrics.mean_absolute_percentage_error(Y_true, Y_predicted)
    return {"R2": R2_score, "MSE": MSE_score, "MAPE": MAPE_score}
```

## Train, test split

```python
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(
    X, Y, test_size=0.2, random_state=0
)
```

## Pairs plot

```python
def pairs_plot(df, line_length=5, kind="scatter"):
    for i in range(0, len(df.columns), line_length):
        sns.pairplot(
            df,
            x_vars=df.columns[i : min(i + line_length, len(df.columns))],
            y_vars="MEDV",
            kind=kind,
        )
```

### Train set

```python
df_train = X_train.copy()
df_train["MEDV"] = Y_train
pairs_plot(df_train)
```

### Test set

```python
df_test = X_test.copy()
df_test["MEDV"] = Y_test
pairs_plot(df_test)
```

## Seed

```python
torch.manual_seed(0)
np.random.seed(0)
random.seed(0)
```

## Model

```python
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(13, 32),
            nn.Tanh(),
            nn.Linear(32, 1),
        )

    def forward(self, x):
        return self.layers(x)


mlp = MLP()
```

## Dataloader

```python
class BostonDataset(torch.utils.data.Dataset):
    def __init__(self, X, Y):
        if not torch.is_tensor(X) and not torch.is_tensor(Y):
            self.X = torch.from_numpy(np.array(X))
            self.Y = torch.from_numpy(np.array(Y))

    def __len__(self):
        return len(self.X)

    def __getitem__(self, i):
        return self.X[i], self.Y[i]


dataset = BostonDataset(X_train, Y_train)
trainloader = torch.utils.data.DataLoader(dataset, batch_size=10, shuffle=True)
```

## Train

```python
n_epochs = 500

loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(mlp.parameters())

for epoch in range(n_epochs):
    total_loss = 0.0
    for data in trainloader:
        inputs, targets = data
        inputs, targets = inputs.float(), targets.float()
        targets = targets.reshape((targets.shape[0], 1))

        # Zero the gradients
        optimizer.zero_grad()

        # Perform forward pass
        outputs = mlp(inputs)

        # Compute loss
        loss = loss_function(outputs, targets)
        total_loss += loss.item()

        # Perform backward pass
        loss.backward()

        # Perform optimization
        optimizer.step()
    print(f"Training epoch=={epoch}: loss=={total_loss}")
```

## Evaluating

```python
df_result = pd.DataFrame()
df_result["Metric"] = performance_evaluate(
    np.array([0, 0, 0]), np.array([0, 0, 0])
).keys()

inputs = torch.from_numpy(np.array(X_train)).float()
outputs = mlp(inputs)
df_result["Train"] = performance_evaluate(outputs.detach().numpy(), Y_train).values()

inputs = torch.from_numpy(np.array(X_test)).float()
outputs = mlp(inputs)
df_result["Test"] = performance_evaluate(outputs.detach().numpy(), Y_test).values()

display(df_result)
```

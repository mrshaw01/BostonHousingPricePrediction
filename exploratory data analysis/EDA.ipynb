## Heatmap

```python
sns.heatmap(data=df.corr(), annot=True)
```

## Pairs plot

```python
def pairs_plot(df, line_length=5, kind="scatter"):
    for i in range(0, len(df.columns), line_length):
        sns.pairplot(
            df,
            x_vars=df.columns[i : min(i + line_length, len(df.columns))],
            y_vars="MEDV",
            kind=kind,
        )
```

```python
pairs_plot(df, line_length=5, kind="scatter")
```

```python
pairs_plot(df, line_length=5, kind="kde")
```

## Remove Outliers

```python
df = df[df["MEDV"] != 50]
df = df[df["RM"] >= 4]
df = df[df["RM"] <= 8.4]

display(df)
sns.heatmap(data=df.corr(), annot=True)
pairs_plot(df, line_length=5, kind="scatter")
pairs_plot(df, line_length=5, kind="kde")
```

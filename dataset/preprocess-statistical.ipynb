# Preprocess dataset by statistical knowledge

### Load dataset from csv file

```python
%load_ext nb_black
```

```python
import pandas as pd
import seaborn as sn
```

```python
# Load processed dataset from csv file
df = pd.read_csv("VN_housing_dataset_processed_expert_knowledge.csv")

print(f"{len(df)} instances")
display(df.head())
```

### Plot relationship between continuous variables & target variable
Before remove outliers

```python
columns = ["floor_num", "bedroom_num", "area", "length", "width", "price/m2"]
sn.pairplot(df, x_vars=columns, y_vars="price/m2")
```

### Remove outliers: out-of-distribution data instance

```python
df = df[df["area"] < 250]
df = df[df["length"] < 32]
df = df[df["width"] < 18]
df = df[df["price/m2"] < 230]
```

### Plot relationship between continuous variables & target variable
After remove outliers

```python
columns = ["floor_num", "bedroom_num", "area", "length", "width", "price/m2"]
sn.pairplot(df, x_vars=columns, y_vars="price/m2")
```

### Export processed dataset to csv file

```python
df.to_csv("VN_housing_dataset_processed_statistical.csv", index=False)
```
